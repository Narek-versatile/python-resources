Task 2: The Polymorphic Loop
Create a list containing one Bitcoin object and one Ethereum object.

Write a simple loop that iterates through the list and prints the transaction fee for an amount of 1000.

Check: Does your code care which specific coin it's looking at, or does it just call the method?

```python
class Coin(ABC):
    @abstractmethod
    def price(self):
        ...

class Bitcoin(Coin):
    def price(self):
        return 72300 #couldnt find a free public api

class Ethereum(Coin):
    def price(self):
        return 2200

def coin_price(coin):
    print(coin.price())

for i in [Bitcoin(), Ethereum()]:
    coin_price(i)
    
```
